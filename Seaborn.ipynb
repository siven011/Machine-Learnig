Seaborn:

  - Data Visualization library

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Note : Seaborn has some built-in datasets

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#setting a theme for the plots
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style ='smoker',size='size')
```

```python
# loading the iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

Scatter plot

```python
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

```python
sns.scatterplot(x='sepal_length',y='petal_width',hue='species',data=iris)
```

```python
# loadding the titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

Count plot


```python
sns.countplot(data=titanic, x = 'class')
```

Bar plot

```python
sns.barplot(data=titanic, x ='sex', y='survived',hue ='class')
```

```python
# house price dataset
from sklearn.datasets import load_boston
house_boston = load_boston()
house = pd.DataFrame(house_boston.data,columns=house_boston.feature_names)
house['PRICE'] = house_boston.target
```

```python
house.head()
```

Distribution plot

```python
sns.distplot(house['PRICE'])
```

Correlation 

Heat Map


```python
correlation = house.corr()
```

```python
# Constructing a Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt = '.3f', annot=True, annot_kws={'size':8}, cmap='Reds')
```
