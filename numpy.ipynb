Numpy Arrays


```python
#list
list1 = [1,3,4,5,6]
print(list1)
type(list1)
```

```python
# CREATING A 1D ARRAY
np_array = np.array([1,3,4,5,6])
print(np_array)
type(np_array)
np_array.shape
```

```python
#CREATING A 2D ARRAY
b = np.array([(1,3,4),(4,5,6),(7,8,9)])
print(b)
b.shape
```

```python
c = np.array([1,3,4],dtype=float)
print(c)
```

Inital Placeholders in numpy arrays

```python
# CREATE NUMPY ARRAY OF ZEROS
x = np.zeros((4,5))
print(x)
```

```python
#NUMPY ARRAY OF ONES
y = np.ones((3,4))
print(y)
```

```python
# ARRAY OF PARTICULAR VALUE
z = np.full((3,4),5)
print(z)
```

```python
# IDENTITY MATRIX
a = np.eye(3) 
print(a)
```

```python
# ARRAY WITH RANDOM VALUES
b = np.random.random((3,4))
print(b)
```

```python
# ARRAY WITH RANDOM VALUES (INTEGERS)
b = np.random.randint(10,100,(3,3))
print(b)
```

```python
# ARRAY OF EVENLY SPACED VALUES ---> SPECIFYING THE NUMBER OF VALUES REQUIRED
d = np.linspace(10,30,6)
print(d)
```

```python
# ARRAY OF EVENLY SPACED VALUES ---> SPECIFYING THE STEP
d = np.arange(10,30,5)
print(d)
```

```python
# CONVERTING LIST TO NUMPY ARRAY
ls = [1,34,5,6,8]
npArray = np.asarray(ls)
print(npArray)
```

Analyzing a numpy array

```python
c = np.random.randint(10,90,(5,5))
print(c)
print(c.shape)
print(c.ndim)
#no. of elements
print(c.size)
#datatype
print(c.dtype)
```

**Mathematical operations on a nparray**

```python
#list concatenation
ls1 = [1,3,4]
ls3 = [3,4,5,6]
print(ls1 + ls3)
```

```python
a = np.random.randint(0,10,(3,3))
b = np.random.randint(0,30,(3,3))
print(a)
print(b)
```

```python
#operation on elements
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
# or
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

Array Manipulation

```python
# transpose
print(a)
trans = np.transpose(a)
print(trans)
#or
trans3 = a.T
print(trans3)
```

```python
# reshape an array
a = np.random.randint(0,10,(3,4))
print(a)
b = a.reshape(6,2)
print(b)
```
