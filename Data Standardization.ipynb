Data Standardization:
- The process of standardizing the data to a common format and common range

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# loading the dataset
dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(dataset)
```

```python
# loading data to pandas DataFrame
df = pd.DataFrame(dataset.data, columns = dataset.feature_names)
```

```python
df.head()
```

```python
df.shape
```

```python
X = df
Y = dataset.target
```

```python
print(Y)
```

Spilitting the data into training data and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Standardize the data

```python
print(dataset.data.std()) # If standard deviation > 1 then the data is not in range and deviates a lot
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_standardized = scaler.transform(X_train)
```

```python
print(X_train_standardized)
```

```python
X_test_standardized = scaler.transform(X_test)
```

```python
print(X_test_standardized.std())
```

```python
print(X_train_standardized.std())
```
