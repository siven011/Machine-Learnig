# Matplotlip

It is a plotting library for the Python programming language and its numerical mathematical extension numpy. It provides an object oriented API for embedding plots into applications using general purpose GUI toolkits like Tkinter, wxPython, Qt ot GTK+.

Some of the major pros of matplotlib are:

- Generally easy to get started for simple plots
- Support for custom labels and texts
- Great control of every element in the figure
- High quality output in many formats
- Very customizable in general

```python
import matplotlib.pyplot as plt

%matplotlib inline  # To avoid using 'plt.show()' in every plots
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)

a = np.arange(40,50)
b = np.arange(50,60)
```

```python
## plotting using matplotlib

## plt scatter

plt.scatter(x,y,c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
y = x*x
```

```python
#plt plot

plt.plot(x,y,'r--o', linewidth=0.9,markersize=4)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Diagram')
```

```python
# Creating subplots (multiple plots in one diagram)
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g-*')
plt.subplot(2,2,3)
plt.plot(x,y,'b--x')
plt.subplot(2,2,4)
plt.plot(x,y,'yo--')
```

```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)

plt.plot(x,y,'g--')
plt.title('Sine wave')
```

```python
## plt bar

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2, color = 'r')
plt.title('Bar Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

# Histograms



```python
## plt hist

a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27,0,0,0,0])
plt.hist(a)
plt.title('Histogram')
```

# Box Plot using Matplotlib

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

## rectangular box plot (generally seaborn is used for box plot)

plt.boxplot(data, vert= True, patch_artist= True);
```

# Pie Chart

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [295,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode =(0.08,0.01,0.01,0.01) # explode first slice

plt.pie(sizes, labels=labels, explode = explode, colors=colors, autopct = '%1.1f%%', shadow = True)

plt.axis('equal')
```
